Basic Line Plots

```python
#import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# set up matplotlib for inline plotting (only for jupyter notebooks)
# if using python scripts then use plt.show()
%matplotlib inline

#plot style
plt.style.use('default')
```

```python
#Creating Sample Data
x=[1,2,3,4,5,6]
y=[2,4,6,8,10,12]
#create the plot
plt.plot(x,y)
plt.show()
print("This is a basic line connecting points (1,2), (2,4)...")
```

```python
#create sample data
X = np.linspace(0, 2*np.pi, 100) # 100 points from 0 to 2π
y = np.sin(X)

#creating the plot
plt.plot(X, y)
plt.show()

print("This plot shows the sin function from 0 to 2π")
```

Customizing plot styles

```python
# Create sample data
X = np.linspace(0, 2*np.pi, 100) # 100 points from 0 to 2π
y = np.sin(X)

# create the plot
plt.figure(figsize=(10,4)) # figure size
plt.plot(X, y, color="red", linewidth=3, linestyle='-', label="sin(x)")
plt.title("Sine Wave Function", fontsize=16, fontweight='bold') # plot title
plt.xlabel("x (radians)", fontsize=16) # X axis label
plt.ylabel("sin(x)", fontsize=16) # y axis label
plt.grid(True, alpha=0.3) # addition the grid
plt.legend() # Display chart labels
plt.show()
```

```python
# Create sample data
X = np.linspace(0, 2*np.pi, 100) # 100 points from 0 to 2π
y = np.sin(X)
y1 = np.cos(X)
y2 = np.tan(X)
y3 = np.sin(X) * np.cos(X)

# create the plot
plt.figure(figsize=(10,4)) # figure size
plt.plot(X, y, color="red", linewidth=3, linestyle='-', label="sin(x)")
plt.plot(X, y1, color="blue", linewidth=3, linestyle='-.', label="cos(x)")
#plt.plot(X, y2, color="green", linewidth=3, linestyle='--', label="tan(x)")
plt.plot(X, y3, color="orange", linewidth=3, linestyle=':', label="sin(x) * cos(x)")
plt.title("Sine Wave Function", fontsize=16, fontweight='bold') # plot title
plt.xlabel("x (radians)", fontsize=16) # X axis label
plt.ylabel("y", fontsize=16) # y axis label
plt.grid(True, alpha=0.3) # addition the grid
plt.legend() # Display chart labels
plt.show()
```

Object oriented plotting

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# subplot 1: sine function
axes[0, 0].plot(X, y, 'r-')
axes[0, 0].set_title('Sine Function')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('sin(x)')
axes[0, 0].grid(True)

# subplot 2: Cosine function
axes[0, 1].plot(X, y1, 'g-')
axes[0, 1].set_title('Cosine Function')
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('cos(x)')
axes[0, 1].grid(True)

# subplot 3: Tan function
axes[1, 0].plot(X, y2, 'b-')
axes[1, 0].set_title('Tan Function')
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('tan(x)')
axes[1, 0].grid(True)

# subplot 4: sin(x) * cos(x)
axes[1, 1].plot(X, y3, 'r-')
axes[1, 1].set_title('sin(x) * cos(x)')
axes[1, 1].set_xlabel('x')
axes[1, 1].set_ylabel('sin(x) * cos(x)')
axes[1, 1].grid(True)
```

Different plot types


*   Bar charts
*  Scatter plots
*   Pie chart












```python
# Bar Charts
categories = ['A', "B", "C", "D", "E"]
values = [23, 17, 35, 29, 12]

plt.figure(figsize=(10,6)) # figure size
bars = plt.bar(categories, values, color=['red', 'green', 'blue', 'orange', 'purple'])
plt.title("Bar Chart Example", fontsize=16, fontweight='bold') # plot title
plt.xlabel("Categories", fontsize=16) # X axis label
plt.ylabel("Values", fontsize=16) # y axis label

# add value labels on top of the bars
for bar, value in zip(bars, values):
  plt.text(bar.get_x() + bar.get_width() / 2, # Center the text on the bar
           bar.get_height() + 0.5,  # Add a small offset above the bar
           str(value), # Convert value to string for display
           ha='center', # Horizontal alignment
           va='bottom' # Vertical Alignment
           )

plt.grid(True, alpha=0.3, axis='y') # addition the grid
plt.show()
```
